```python
import numpy as np

# Function to compute Levenshtein distance matrix
def findDis(a, b):
    alen = len(a) + 1
    blen = len(b) + 1
    m = [[0 for i in range(alen)] for j in range(blen)]
    
    for i in range(1, alen):
        m[0][i] = i
    for j in range(1, blen):
        m[j][0] = j
    
    for j in range(1, alen):
        for i in range(1, blen):
            if a[j - 1] == b[i - 1]:
                f = 0
            else:
                f = 1
            one = m[i - 1][j - 1] + f
            two = m[i - 1][j] + 1
            three = m[i][j - 1] + 1
            m[i][j] = min(one, two, three)
    
    return m

# Function to print the matrix and edit distance
def printvals(a, b):
    c = findDis(a, b)
    l = a.split()
    
    print("\t\t", end=' ')
    for k in range(len(a)):
        print(a[k], "\t", end='')
    print("\n")
    
    for i in range(len(c)):
        if i == 0:
            print('\t', end='')
        if i > 0:
            print(b[i - 1], '\t', end='')
        for j in range(len(c[0])):
            print(c[i][j], '\t', end='')
        print("\n")
    
    print("Edit Distance=", c[-1][-1])
    return

# Example strings
string1 = "kitten"
string2 = "sitting"

# Print the Levenshtein distance matrix step by step
printvals(string1, string2)
```
